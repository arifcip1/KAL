# Menyelesaikan Sistem Persamaan Linear 3 Variabel dengan Eliminasi Gauss
Kita akan mempelajari tiga tipe solusi untuk sistem persamaan linear tiga variabel:
- Tidak memiliki solusi
- Memiliki satu solusi
- Memiliki solusi tak hingga

Setiap kasus akan diselesaikan menggunakan metode **Eliminasi Gauss** dan divisualisasikan matriks augmentasinya.

## 📌 Kasus 1: Sistem Tanpa Solusi

```
import numpy as np
from sympy import Matrix

# Sistem persamaan
A1 = np.array([[1, 1, 1],
               [1, 1, 1],
               [1, 1, 1]])
b1 = np.array([[4], [6], [8]])

aug1 = np.hstack([A1, b1])
M1 = Matrix(aug1)
display(M1.rref())  # Bentuk eselon baris tereduksi

print("Dari hasil eliminasi, terdapat baris seperti [0 0 0 | k] dengan k ≠ 0. Artinya sistem tidak konsisten dan tidak memiliki solusi.")
```

## 📌 Kasus 2: Sistem dengan Satu Solusi

```
# Sistem persamaan dengan satu solusi
A2 = np.array([[1, 1, 1],
               [2, -1, 3],
               [-1, 4, -1]])
b2 = np.array([[6], [14], [-2]])

aug2 = np.hstack([A2, b2])
M2 = Matrix(aug2)
rref2, _ = M2.rref()
display(rref2)

# Substitusi balik untuk solusi
from sympy import symbols, Eq, solve
x, y, z = symbols('x y z')

# Ambil persamaan dari hasil eliminasi
eq3 = Eq((5/3)*z, 22/3)
sol_z = solve(eq3, z)[0]

eq2 = Eq(-3*y + sol_z, 2)
sol_y = solve(eq2, y)[0]

eq1 = Eq(x + sol_y + sol_z, 6)
sol_x = solve(eq1, x)[0]

print(f"Solusi: x = {sol_x}, y = {sol_y}, z = {sol_z}")
```

## 📌 Kasus 3: Sistem dengan Solusi Tak Hingga

```
# Sistem dengan solusi tak hingga
A3 = np.array([[1, 1, 1],
               [2, 2, 2],
               [3, 3, 3]])
b3 = np.array([[6], [12], [18]])

aug3 = np.hstack([A3, b3])
M3 = Matrix(aug3)
display(M3.rref())

print("Hanya satu persamaan yang independen. Artinya sistem memiliki solusi tak hingga yang dapat dituliskan dalam bentuk parametrik:")

y, z = symbols('y z')
x_param = 6 - y - z
print(f"x = {x_param}, y = y, z = z")
```

## 📊 Kesimpulan
1. Sistem yang tidak konsisten menghasilkan **tidak ada solusi**.
2. Sistem dengan semua persamaan independen dan tidak bertentangan menghasilkan **satu solusi unik**.
3. Sistem yang memiliki redundansi (persamaan tidak independen) menghasilkan **solusi tak hingga**.

Metode Eliminasi Gauss membantu kita mengidentifikasi jenis solusi tersebut secara sistematis.
